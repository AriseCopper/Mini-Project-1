```python
%gui tk
import tkinter as tk
import random
secret_number = random.randint(1, 100)

def check_guess():
    try:
        guess = int(entry.get())
    except ValueError:
        result_label.config(text="‚ùå Please enter a valid number!", fg="red", bg="#FFCCCC")
        return
    
    if guess < secret_number:
        result_label.config(text="üìâ Too Low! Try again.", fg="orange", bg="#FFF3CD")
    elif guess > secret_number:
        result_label.config(text="üìà Too High! Try again.", fg="orange", bg="#FFF3CD")
    else:
        result_label.config(text=f"üéâ Correct! The number was {secret_number}", fg="white", bg="#28A745")

def reset_game():
    global secret_number
    secret_number = random.randint(1, 100)
    entry.delete(0, tk.END)
    result_label.config(text="üîÑ New game started! Guess a number.", fg="blue", bg="#CCE5FF")

root = tk.Tk()
root.title("üéØ Number Guessing Game")
root.geometry("450x300")
root.config(bg="#222831")  

title_label = tk.Label(
    root, 
    text="üéØ Guess the Number (1-100)", 
    font=("Arial", 18, "bold"), 
    fg="#FFD369", 
    bg="#222831"
)
title_label.pack(pady=15)

entry = tk.Entry(root, font=("Arial", 16), justify="center", width=8, bg="#EEEEEE")
entry.pack(pady=10)

btn_frame = tk.Frame(root, bg="#222831")
btn_frame.pack(pady=5)

check_button = tk.Button(
    btn_frame, text="Check Guess", font=("Arial", 12, "bold"), 
    command=check_guess, bg="#00ADB5", fg="white", width=12
)
check_button.grid(row=0, column=0, padx=5)

reset_button = tk.Button(
    btn_frame, text="Reset Game", font=("Arial", 12, "bold"), 
    command=reset_game, bg="#FF5722", fg="white", width=12
)
reset_button.grid(row=0, column=1, padx=5)

result_label = tk.Label(root, text="", font=("Arial", 14, "bold"), bg="#222831", fg="white", wraplength=400, justify="center")
result_label.pack(pady=20)

root.mainloop()
```
